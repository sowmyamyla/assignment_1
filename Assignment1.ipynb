```python
#Question 1
#a.1,2,3
mat = np.random.randint(1, 21, size=15)
arr = mat.reshape(3, 5)
print("Array shape:", arr.shape)
arr[np.arange(arr.shape[0]), arr.argmax(axis=1)]
print("Modified array:\n", arr)
```

```python
#1.b
arr = np.zeros((4, 3), dtype=np.int32)
print("Array:\n", arr)
print("Shape:", arr.shape)
print("Type:", type(arr))
print("Data type:", arr.dtype)
```

```python
arr = np.array([[3, -2], [1, 0]])
eigenvalues, eigenvectors = np.linalg.eig(arr)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

```python
#1.c
arr = np.array([[0, 1, 2], [3, 4, 5]])
sum_diag = np.trace(arr)
print("Array:\n", arr)
print("Sum of diagonal elements:", sum_diag)
```

```python
#1.d
arr = np.array([[1, 2], [3, 4], [5, 6]])
reshape_1 = np.reshape(arr, (2, 3))
reshape_2 = np.reshape(arr, (3, 2))
print("Original array:\n", arr)
print("Reshaped to 3x2:\n", reshape_2)
print("Reshaped to 2x3:\n", reshape_1)
```
